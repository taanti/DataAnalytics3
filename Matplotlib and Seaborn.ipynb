# Data Visualization Using Python

Python offers several plotting libraries, namely Matplotlib, Seaborn and many other such data visualization packages with different features for creating 
informative, customized, and appealing plots to present data in the most simple and effective way.

## Matplotlib and Seaborn

Matplotlib and Seaborn are python libraries that are used for data visualization. They have inbuilt modules for plotting different graphs. While Matplotlib is used to embed graphs into applications, Seaborn is primarily used for statistical graphs.

But when should we use either of the two? Let's understand this with the help of a comparative analysis. Below are the comparison between Python's two well-known visualization packages Matplotlib and Seaborn,

### Matplotlib

- It is used for basic graph plotting like line charts. bar graphs. etc.
- It mainly works with datasets and arrays.
- Matplotlib acts productively with data arrays and frames. It regards the aces and figures as objects
- Matplotlib is more customizable and pairs well with Pandas and Numpy for Exploratory Data Analysis.

### Seaborn
- It is mainly used for statistics visualization and can perform complex visualizations with fewer commands.
- It works with entire datasets.
- Seaborn is considerably more organized and functional than Matplotlib and treats the entire dataset as a solitary unit.
- Seaborn has more inbuilt themes and is mainly used for statistical analysis.

## Line Chart
A Line chart is a graph that represents information as a series of data points connected by a straight line. In line charts, each data point or marker is plotted and connected with a line or curve.

Lets consider the apple yield (tons per hectare) in Kanto. Lefs plot a line graph using this data and see how the yield Of apples changes over time. We are going to both plot by using Matplotlib and Seabom as follows:

### Using Matplotlib

We are using random data points to represent the yeild of apples

```python
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95] #Ton per Hectare
```

After we define the data points of apple yields, we are going to plot its Line Chart using Matplotlib.

```python
plt.plot(yield_apples)
```

To better understand the graph and its purpose. we can add the x-axis values too.

```python
years = [2017, 2018, 2019, 2020, 2021, 2022]
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95]
```

Since we have the 'years' variable as the x-axis, we need to type it before the yield _ apples variable below

```python
plt.plot(years, yield_apples) #plt.plot(x,y)
```

To plot multiple datasets on the same graph, just use the plt.plot function once for each dataset Let's use this to compare the yields Of apples vs. oranges on 
the same graph.

```python
apples = [0.89, 0.91, 0.93, 0.97, 0.986, 0.943]
tangerines = [0.79, 0.61, 0.93, 0.97, 0.962, 0.931]
```

```python
plt.plot(years, apples)
plt.plot(years, tangerines)
plt.xlabel('Year')
plt.ylabel('Yield (Tons per Hectare)')
```

We can add a legend Which tells us What each line in our graph means. TO understand What we are plotting. we can add a title to our graph.

```python
plt.plot(years, apples)
plt.plot(years, tangerines)

plt.xlabel('Year')
plt.ylabel('Yield (Tons per Hectare)')

plt.title('Crop Yields in Bandung')
plt.legend(['Apples', 'Tangerines'])
```

## Bar Chart

When you have categorical data, you can represent it with a bar graph. A bar graph plots data with the help of bars. which represent value on the y-axis and category on the x-axis. Bar graphs use bars with varying heights to show the data which belongs to a specific category.

### Using Matplotlib

```python
plt.bar(years, tangerines) #plt,bar(x,y)

plt.xlabel('Year')
plt.ylabel('sales')

plt.title('Crop Yields in Kanto')
```

```python
plt.bar(years, apples)
plt.bar(years, tangerines, bottom=apples)
```

#### **Another Example** :

```python
import numpy as np
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
apples = [100, 120, 110, 105, 115]  # Penjualan apel per bulan
oranges = [80, 85, 90, 95, 100]      # Penjualan jeruk per bulan
grapes = [60, 70, 75, 80, 85]        # Penjualan anggur per bulan
```

```python
plt.bar(months, grapes) #plt,bar(x,y)

plt.xlabel('Month')
plt.ylabel('Fruit Sales per Month')

plt.title('Grapes Sales per Month')
```

```python
# Menghitung total penjualan untuk setiap bulan
total_sales = np.add(apples, oranges)

plt.bar(months, apples, label='Apples')
plt.bar(months, oranges, bottom=apples, label='Oranges')
plt.bar(months, grapes, bottom=total_sales, label='Grapes')

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Fruit Sales per Month')

plt.legend()
plt.show()
```

### Using Seaborn

Let's use the tips dataset in Seaborn next. The dataset consists of :
1. Information about the sex (gender)
2. Time of day
3. Total bill
4. Tips given by customers visiting the restaurant for a week
   
This Tips Dataset is a dataset provide by seaborn library, so you can directly load the dataset.

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

We can draw a bar chart to visualize how the average bill amount varies across different days of the week. We can do this by computing the day-wise averages and then using plt.bar. The Seaborn library also provides a barplot function that can automatically compute averages.

```python
sns.barplot(x='day',
            y='total_bill',
            data=tips_df)
```

If you want to compare bar plots side-by-side. you can use the hue argument. The comparison will be done based on the third feature specified in this
argument.

```python
sns.barplot(x='day', y='total_bill', hue='sex', data = tips_df)
```

```python
sns.barplot(x='total_bill', y='day', hue='sex', data = tips_df)
```

## Histogram

A Histogram is a bar representation of data that varies over a range. It plots the height of the data belonging to a range along the y-axis and the range along
the x-axis. Histograms are used to plot data over a range of values. They use a bar representation to show the data belonging to each range. Let's again use
the 'Iris' data which contains information about flowers to plot histograms.

The 'iris' dataset is a dataset provided by Seaborn which is completely numerical.

```python
iris_df = sns.load_dataset('iris')
iris_df.head()
```

 ### Using Matplotlib

We are going to plot a histogram from the 'Sepal Width' column.

```python
iris_df.sepal_width
```

```python
plt.title("Distribustion of Sepal Width")
plt.hist(iris_df.sepal_width);
```

We can control the number or size of bins too.

```python
#specifying the number of bins
plt.hist(iris_df.sepal_width, bins=10);
```

We can change the number and size of bins using numpy too.

```python
#specifying the number of bins
plt.hist(iris_df.sepal_width, bins=np.arange(2, 5, 0.25));
```

We can create bins of unequal size too.

```python
#specifying the number of bins use Matplotlib
plt.hist(iris_df.sepal_width, bins=[1,3,4,4.5]);
```

Similar to line charts, we can draw multiple histograms in a single chart. We can reduce each histogram's opacity so that one histogram's bars don't hide the
other. Let's draw separate histograms for each species of flowers.
